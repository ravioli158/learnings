### 1. Write a short Python function, is_multiple(n, m), that takes two integer values and retuns True if n is a multiple of m, that is , n = mi for some integer i, and False otherwise.

```python
def is_multiple(n, m):
    try:
        return n % m == 0 and n / m >= 1
    except Exception as e:
        print(e)


print(is_multiple(4, 2))
print(is_multiple(0, 2))
print(is_multiple(2, 0))
```

```python
"""
R-1.2 Write a short Python function, is even(k), that takes an integer value and
returns True if k is even, and False otherwise. However, your function
cannot use the multiplication, modulo, or division operators.
"""


def is_even(n):

    for i in range(2, n + 1, 2):
        print(i)
        if i == n:
            return True
        elif i > n:
            return False


print(is_even(8))
print(is_even(13))
```

```python
"""
R-1.3 Write a short Python function, minmax(data), that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution.
"""

pos_inf = float("inf")
neg_inf = float("-inf")


def minmax(data):
    """
    time compexity: O(n), where n is length of data
    """
    min = pos_inf
    max = neg_inf
    for num in data:
        if num > max:
            max = num
        if num < min:
            min = num
    return min, max


print(minmax([12, 67, -4, 98]))
```

```python
"""
R-1.4 Write a short Python function that takes a positive integer n and returns
the sum of the squares of all the positive integers smaller than n.
"""


def sum_squares(n):
    sum = 0
    i = 1
    while i < n:
        sum += i * i
        i += 1
    return sum


print(sum_squares(4))
```

```python
"""
R-1.6 Write a short Python function that takes a positive integer n and returns
the sum of the squares of all the odd positive integers smaller than n.
"""


def sum_odd_pos(n):
    if not isinstance(n, int) or n <= 0:
        raise TypeError("n should be positive integer")
    result = 0
    for i in range(1, n):
        if i % 2 != 0:
            result += i * i
    return result


print(sum_odd_pos(10))
```
