```python
# Sets
eatable = {'mango', 'apple', 'potato', 'milk'}
fruits = {'mango', 'apple'}
drinks = {'milk','juice'}
mango = 'mango'
print(f"Is {mango} in {eatable}: {mango in eatable}")
print(f"is drinks: {drinks} subset of eatables: {eatable}: {drinks < eatable}")
print(f"is drinks:{drinks} proper subset of eatables: {eatable}: {drinks <= eatable}")
#atleast one is there
print(bool(drinks & eatable))
```

```python
#Dictionaries
languages = {'en':'english','in':'hindi'}
test_dict = {
    'en': 'english',
    'in':'hindi'
}
print(languages)
print('en' in languages)
print(test_dict == languages)
print(test_dict is languages)
```

```python
# Iterator and Generator
# * iterator: object that manages an iteration through a series of values.
data = [1,23]
# next(data), data is iterable but it is not an iterator object
iterator_data = iter(data)
while True:
    try:
        print(next(iterator_data))
    except:
        break
ran = range(5)
i = iter(ran)
print(next(i))
```
