# 1. Write a short Python function, is_multiple(n, m), that takes two integer values and retuns True if n is a multiple of m, that is , n = mi for some integer i, and False otherwise.

```python
def is_multiple(n, m):
    try:
        return n % m == 0 and n / m >= 1
    except Exception as e:
        print(e)
print(is_multiple(4,2))
print(is_multiple(0,2))
print(is_multiple(2,0))
```

```python
def is_even(n):
    for i in range(2, n+1, 2 ):
        print(i)
        if i == n:
            return True
        elif i > n:
            return False
        
print(is_even(8))
print(is_even(13))
```
