```python
# Tuples are immutable sequences in python, used to store items that cannot be changed after creation. 
fruits = ('mango', 'banana', 'apple')
# * is unpacking operator before an iterable(like a tuple)
food = (*fruits, 'roti')
perishable_items = (*food, 'milk')
print(perishable_items)

# Named Tuples
from collections import namedtuple
Person = namedtuple('Person', ['name', 'age', 'city'])
ravi = Person('Ravi', 22, 'Bengaluru')
print(f"{ravi.name} is {ravi.age} years old and lives in {ravi.city}")


# Tuples allow unpacking
name, age, city = ravi
print(name,age,city)

# named tuples are immutable


# Dataclass
```

```python
# named tuple
from collections import namedtuple
# Card = namedtuple('Card', ['suit', 'rank'])
# beer_card = Card('7','diamond')
# print(beer_card)

from dataclasses import dataclass
from enum import Enum

class Suit(Enum):
    Diamond = 'Diamond'
    Spade = 'Spade'
    Heart = 'Heart'
    Club = 'Club'
    def __str__(self):
        return self.value

    
Rank = Enum('Rank', [str(n) for n in range(2,11)] + ['A', 'J','Q', 'K'])

@dataclass(frozen=True)
class Card:
    suit: Suit 
    rank: Rank
```
